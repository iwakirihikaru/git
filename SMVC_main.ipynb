```python
class GarbageDayReminder(object):
    
    def __init__(self):
        self.scheduler = scheduler(time.time, time.sleep)

    def check1_timer(self): #実行内容
        print("check1実行")
        num=0
        check1(numbers)
    
    def last_timer(self): #実行内容
        print("last実行")
        num=0
        last(numbers)
    
    def sikomi3_timer(self): #実行内容
        print("sikomi3実行")
        num=0
        sikomi3(numbers)
        
    def wifi_timer(self):
        print(now)
        wifi_entyo(numbers)

    def specified_time(self):
        """
        スケジューラーに予定を登録し実行する。
        """
        now = datetime.now()
        run1 = now.replace(hour=9, minute=0, second=0, microsecond=0)
        run2 = now.replace(hour=22, minute=59, second=10, microsecond=0)
        run3 = now.replace(hour=14, minute=50, second=0, microsecond=0)
        run4 = now.replace(hour=15, minute=5, second=10, microsecond=0)
        wifi = now.replace(hour=8, minute=10, second=0, microsecond=0)
        wifi2 = now.replace(hour=8, minute=40, second=0, microsecond=0)

        
        if now > run1:
            # スケジュール登録時に時間が経過していたら次の日に
            run1 += timedelta(days=1)
            pass
        if now > run2:
            # スケジュール登録時に時間が経過していたら次の日に
            run2 += timedelta(days=1)
            pass     
        if now > run3:
            # スケジュール登録時に時間が経過していたら次の日に
            run3 += timedelta(days=1)
            pass
        if now > run4:
            # スケジュール登録時に時間が経過していたら次の日に
            run4 += timedelta(days=1)
            pass
        
        print(f"仕込み実行予定時刻:{run1}")
        print(f"仕込み実行予定時刻:{run2}")
        print(f"1％値上がり購入仕込み実行予定時刻:{run3}")        
        print(f"大引け後ラスト実行予定時刻:{run4}")

        self.scheduler.enterabs(run1.timestamp(), 1, self.check1_timer) #時間、優先順位、呼び出し関数　timestamp..unix時間に変換らしい 関数に()が付くとなぜかうまく実行しない。注意！
        self.scheduler.enterabs(run3.timestamp(), 1, self.sikomi3_timer) #sikomi3
        self.scheduler.enterabs(run4.timestamp(), 1, self.last_timer) #last
        self.scheduler.enterabs(wifi.timestamp(), 1, self.wifi_timer) #last
        self.scheduler.enterabs(wifi2.timestamp(), 1, self.wifi_timer) #last

        self.scheduler.run()

def main() -> None:
    # GarbageDayReminderをインスタンス化
    g = GarbageDayReminder()
    # スレッドで指定時間動作関数を動かす
    t = Thread(target=g.specified_time)
    t.start()


if __name__ == '__main__': #http://azuuun-memorandum.hatenablog.com/entry/2015/05/09/002549 参照
    main()
```
